```python
!pip install statsmodels
import statsmodels.api as sm
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
airline_costs = pd.read_csv("Downloads/airline_costs.csv")
```

```python
airline_costs.head()
```

```python

X = airline_costs[['DailyFlightTime', 'FlightLength']]
y = airline_costs['CustomersServed']

x1 = sm.add_constant(X)
model = sm.OLS(y, x1).fit()
model.summary()
```

```python
x2 = airline_costs[['CustomersServed']]
y2 = airline_costs['TotalAssets']

x2 = sm.add_constant(x2)
model = sm.OLS(y2, x2).fit()
model.summary()
```

The two regression models result in different r squared outputs because when only one independent variable is input into the model versus two, there is a higher proportion of variance that is due to this variable. However, typically the more variables are included the higher the r squared becomes. The first model with two variables could has a lower r-squared. This is possibly because neither variables have a strong relationship with the number of customers on each flight.
